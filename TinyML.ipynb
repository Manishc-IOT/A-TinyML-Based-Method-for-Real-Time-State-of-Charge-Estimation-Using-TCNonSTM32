```python
# ===============================
# 🔹 INSTALL
# ===============================
!pip install tensorflow keras-tcn scikit-learn openpyxl

# ===============================
# 🔹 IMPORTS
# ===============================
import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from tcn import TCN

from google.colab import files

# ===============================
# 🔹 UPLOAD DATASET
# ===============================
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

df = pd.read_excel(file_name)
df.columns = df.columns.str.strip()

# Keep required columns
df = df[["Voltage", "Current", "Temperature"]]

print("Initial data:")
print(df.head())

# ===============================
# 🔥 SOC GENERATION (IMPROVED)
# ===============================
v_min = df["Voltage"].min()
v_max = df["Voltage"].max()

i_mean = df["Current"].mean()
t_mean = df["Temperature"].mean()

# data-driven weights
k_i = 0.05 * (np.std(df["Voltage"]) / np.std(df["Current"]))
k_t = 0.02 * (np.std(df["Voltage"]) / np.std(df["Temperature"]))

soc_list = []
soc_prev = None

for i in range(len(df)):

    v = df["Voltage"].iloc[i]
    i_val = df["Current"].iloc[i]
    t = df["Temperature"].iloc[i]

    # voltage dominates
    soc_v = (v - v_min) / (v_max - v_min) * 100

    # corrections
    delta_i = k_i * (i_val - i_mean)
    delta_t = k_t * (t - t_mean)

    soc_inst = soc_v + delta_i + delta_t

    # bound correction
    if soc_inst > soc_v + 7: soc_inst = soc_v + 7
    if soc_inst < soc_v - 7: soc_inst = soc_v - 7

    # cumulative smoothing
    if soc_prev is None:
        soc = soc_inst
    else:
        soc = 0.8 * soc_prev + 0.2 * soc_inst

    soc_prev = soc
    soc = max(0, min(100, soc))

    soc_list.append(round(soc, 4))

df["SOC"] = soc_list

print("\nSOC generated")

# 🔥 ADD THIS HERE
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df["Voltage"])
plt.xlabel("Time")
plt.ylabel("Voltage (V)")
plt.title("Battery Voltage Discharge Curve")
plt.show()
plt.savefig("voltage_plot.png", dpi=300)


# ===============================
# 🔹 SCALING (STANDARD)
# ===============================
scaler = StandardScaler()
scaled = scaler.fit_transform(df)

print("\nMEAN:", scaler.mean_)
print("STD :", scaler.scale_)

# ===============================
# 🔹 CREATE SEQUENCES
# ===============================
window = 10

X, y = [], []

for i in range(len(scaled) - window):
    X.append(scaled[i:i+window, 0:3])
    y.append(scaled[i+window, 3])

X = np.array(X)
y = np.array(y)

print("\nSequence shape:", X.shape)

# ===============================
# 🔹 TRAIN TEST SPLIT
# ===============================
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# ===============================
# 🔹 ADD NOISE (GENERALIZATION)
# ===============================
X_train = X_train + np.random.normal(0, 0.01, X_train.shape)

# ===============================
# 🔹 BUILD MODEL
# ===============================
model = Sequential([
    tf.keras.layers.Input(shape=(10,3)),
    TCN(nb_filters=32, kernel_size=3, dilations=[1,2,4], dropout_rate=0.1),
    Dense(32, activation="relu"),
    Dense(16, activation="relu"),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse", metrics=["mae"])

model.summary()

# ===============================
# 🔹 TRAIN
# ===============================
early_stop = EarlyStopping(patience=8, restore_best_weights=True)

model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[early_stop],
    verbose=1
)
#result_1
# import numpy as np

# # smoothing (moving average)
# window = 5
# y_pred = model.predict(X_test)

# soc_std = scaler.scale_[3]
# soc_mean = scaler.mean_[3]

# y_pred_real = y_pred * soc_std + soc_mean
# y_test_real = y_test * soc_std + soc_mean
# y_pred_smooth = np.convolve(y_pred_real.flatten(), np.ones(window)/window, mode='same')

# plt.figure()
# plt.plot(y_test_real[:200], label="Actual SoC")
# plt.plot(y_pred_smooth[:200], label="Predicted SoC")
# plt.xlabel("Sample Index")
# plt.ylabel("SoC (%)")
# plt.legend()
# plt.title("Actual vs Predicted SoC")
# plt.show()
# plt.savefig("result_graph.png", dpi=300)
# ===============================
# 🔹 EVALUATE
# ===============================
loss, mae = model.evaluate(X_test, y_test)

soc_std = scaler.scale_[3]
real_mae = mae * soc_std

print("\n==============================")
print("Scaled MAE:", mae)
print("Real SOC Error (%):", real_mae)
print("==============================")

# ===============================
# 🔥 CLASSIFICATION METRICS
# ===============================
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt

# Convert scaled → real values
soc_std = scaler.scale_[3]
soc_mean = scaler.mean_[3]

y_pred = model.predict(X_test)

y_pred_real = y_pred * soc_std + soc_mean
y_test_real = y_test * soc_std + soc_mean

# ===============================
# 🔹 CONVERT TO CLASSES
# ===============================
def soc_to_class(soc):
    if soc < 20:
        return 0   # Low
    elif soc < 80:
        return 1   # Medium
    else:
        return 2   # High

y_pred_class = [soc_to_class(x) for x in y_pred_real.flatten()]
y_test_class = [soc_to_class(x) for x in y_test_real.flatten()]

# ===============================
# 🔹 METRICS
# ===============================
acc = accuracy_score(y_test_class, y_pred_class)
f1 = f1_score(y_test_class, y_pred_class, average='weighted')

print("\nAccuracy:", acc)
print("F1 Score:", f1)

# ===============================
# 🔹 CONFUSION MATRIX
# ===============================
cm = confusion_matrix(y_test_class, y_pred_class)

plt.figure()
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=["Low","Medium","High"],
            yticklabels=["Low","Medium","High"])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (SoC Classification)")
plt.savefig("confusion_matrix.png", dpi=300)
plt.show()

# ===============================
# 🔥 FULL INT8 QUANTIZATION
# ===============================
def representative_data_gen():
    indices = np.random.permutation(len(X_train))
    for i in indices:
        yield [X_train[i:i+1].astype(np.float32)]

converter = tf.lite.TFLiteConverter.from_keras_model(model)

converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.representative_dataset = representative_data_gen
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]

converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8

tflite_model = converter.convert()


# ===============================
# 🔹 SAVE MODEL
# ===============================
with open("soc_model_int8.tflite", "wb") as f:
    f.write(tflite_model)

print("\n✅ INT8 model saved!")

# ===============================
# 🔹 GET SCALE & ZERO POINT
# ===============================
interpreter = tf.lite.Interpreter(model_content=tflite_model)
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

INPUT_SCALE = input_details[0]['quantization'][0]
INPUT_ZERO  = input_details[0]['quantization'][1]

OUTPUT_SCALE = output_details[0]['quantization'][0]
OUTPUT_ZERO  = output_details[0]['quantization'][1]

print("\nINPUT SCALE:", INPUT_SCALE)
print("INPUT ZERO :", INPUT_ZERO)

print("\nOUTPUT SCALE:", OUTPUT_SCALE)
print("OUTPUT ZERO :", OUTPUT_ZERO)

# ===============================
# 🔹 DOWNLOAD
# ===============================
# files.download("soc_model_int8.tflite")
# from google.colab import files
# files.download("result_graph.png")
# files.download("voltage_plot.png")
from google.colab import files
files.download("confusion_matrix.png")
```
